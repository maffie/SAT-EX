### Plot lat-lon data on a world map

```python
import pandas as pd
import matplotlib.pyplot as plt
```

First, import the module with some custom plotting functions:

```python
import lib_plot as lp
```

This module contains two functions: 
- convert_data() to convert a pandas dataframe to a numpy array suitable for plotting with basemap
- basemap_plot() to make the actual plot

use shift+tab to get more information about the arguments of these functions.

Now, read in a dataframe with data that you want to plot:

```python
DF = pd.read_csv("/home/stijndc/Desktop/phd_sdcubber/SAT-EX/results/debugdump/ts_pixels_correlations_climate_NDVI.csv",  
                 header = 0)
```

The dataframe should contain a 'lat' and a 'lon' column, and a column with some values to plot. The variable 'key' should contain the name of the column you want to show on the map.

```python
key = 'MSWEP_lag1'
```

First, we want to convert the dataframe to a plottable array using convert_data()

```python
globalmaparray = lp.convert_data(DF, key=key)
```

We can now use this array to make a plot

```python
fig, ax = plt.subplots(figsize=(20,12))
lp.basemap_plot(fig, ax, globalmaparray, colscheme='Greens', boundaries=[-0.1,0,0.2,0.5,0.8])
plt.show()
```

You can also pass a custom colormap to the customcol argument. Make sure it has the correct lenght (len(boundaries -1).

```python
from matplotlib import colors as colors
cmap = colors.ListedColormap([ 'black','#edf8e9','red','#74c476','#31a354','yellow'])
```

```python
fig, ax = plt.subplots(figsize=(20,12))
lp.basemap_plot(fig, ax, globalmaparray, customcol=cmap, boundaries=[-0.1,0,0.1,0.15,0.2,0.5,0.8])
plt.show()
```

For more info on colors check http://seaborn.pydata.org/tutorial/color_palettes.html
