Exploratory data analysis on T_ERA residuals

```python
import os
import pandas as pd
import xarray as xr

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read in datafile
datadir = ('/media/stijndc/data/datafiles/SAT_EX/netcdf_files_processed/T_ERA/residuals/')
data = xr.open_dataset(os.path.join(datadir, 'T_ERA_residuals_9D.nc'))
data
```

```python
fig, ax = plt.subplots(figsize=(12,6))
data['T'].sel(time='1982-01-01', method='nearest').plot(ax=ax)
```
